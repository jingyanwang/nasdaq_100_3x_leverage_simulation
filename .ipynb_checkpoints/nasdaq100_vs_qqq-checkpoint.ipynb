```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

def fetch_data(start_date, end_date):
    # Define ticker symbols
    qqq_ticker = 'QQQ'
    nasdaq100_ticker = '^NDX'  # Nasdaq-100 Index

    # Fetch daily data
    qqq_data = yf.download(qqq_ticker, start=start_date, end=end_date)
    nasdaq100_data = yf.download(nasdaq100_ticker, start=start_date, end=end_date)

    # Keep only the 'Close' prices
    qqq_close = qqq_data['Close']
    nasdaq100_close = nasdaq100_data['Close']

    # Combine the data into a single DataFrame
    combined_data = pd.DataFrame({
        'QQQ Close': qqq_close,
        'Nasdaq-100 Close': nasdaq100_close
    })

    # Drop rows with NaN values
    combined_data.dropna(inplace=True)

    return combined_data

def plot_ratios(df):
    # Calculate the ratio of QQQ to Nasdaq-100 Index
    df['Ratio'] = df['QQQ Close'] / df['Nasdaq-100 Close']
    
    # Plotting
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df['Ratio'], label='QQQ/Nasdaq-100 Ratio', color='blue')
    plt.title('QQQ to Nasdaq-100 Index Ratio Over Time')
    plt.xlabel('Date')
    plt.ylabel('Ratio')
    plt.axhline(y=1, color='r', linestyle='--', label='Ratio = 1 (Equal Value)')
    plt.legend()
    plt.grid()
    plt.show()

# Define the date range for the data you want to fetch
start_date = '2020-01-01'  # Adjust as needed
end_date = '2024-10-01'    # Adjust as needed

# Fetch the data
df = fetch_data(start_date, end_date)

# Plot the ratios
plot_ratios(df)
```
